# Picking Productivity Report Assignment

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
# Import path of the file to load data and read
df = pd.read_csv('C:/Users/Sha/Desktop/Yasmine/Kapscomoto/Data.csv')
data = pd.DataFrame(df)
data.head(10)
```

```python
# Remove unwanted columns
data=df.drop(['Product','Transaction Type','Account','Created','Edited'], axis=1)

# Rename the 'Transaction Date' column to 'Date'
data.rename(columns={'Transaction Date': 'Date'}, inplace=True)

# Extract the date from the timestamp column
data['Date'] = pd.to_datetime(data['Date']).dt.date

data.head(10)

```

```python
# Extract the aisle from the bin column
import re

Aisles = []

for bin in data['Bin']:
    match = re.search(r"\[(\w)", bin)
    if match:
        aisle = match.group(1)
        Aisles.append(aisle)
    else:
        aisle = bin[0]
        Aisles.append(aisle)

data['Aisle']=Aisles
print(data.head(10))
```

```python
# Add a Column named Count to the data.
data['Count'] = data.groupby(['Date', 'User', 'Bin'])['Qty'].transform('count')
print(data.head(10))
```

## Report By Date

```python
# Group the data by date and calculate count and sum of quantities
Report_By_Date = data.groupby('Date').agg({'Count': 'count', 'Qty': 'sum'}).abs().reset_index()

Report_By_Date = Report_By_Date.style

# Apply CSS styles to bold and center align the row and column headings
Report_By_Date.set_table_styles([
    {'selector': 'th',
     'props': [
         ('text-align', 'center'),
         ('font-weight', 'bold')
     ]},
    {'selector': 'td',
     'props': [
         ('text-align', 'center')
     ]}
])

# Display the report by date
print("\033[1mPicking Productivity Report by Date\033[0m")
print("-----------------------------------")
display(Report_By_Date)
```

## Report By Employee

```python
import numpy as np

Report_By_Employee = pd.pivot_table(data, values=['Count','Qty'], index='User', columns='Date', aggfunc={'Count': 'count', 'Qty': np.sum}, fill_value='')

# Remove decimal places and format as integers
Report_By_Employee = Report_By_Employee.applymap(lambda x: abs(int(x)) if pd.notnull(x) and isinstance(x, float) else x)
Report_By_Employee = Report_By_Employee.style

# Apply CSS styles to bold and center align the row and column headings
Report_By_Employee.set_table_styles([
    {'selector': 'th',
     'props': [
         ('text-align', 'center'),
         ('font-weight', 'bold')
     ]},
    {'selector': 'td',
     'props': [
         ('text-align', 'center')
     ]}
])

# Display the report by Employee
print("\033[1mPicking Productivity Report by Employee\033[0m")
print("---------------------------------------")
display(Report_By_Employee)
```

## Report By Aisle

```python
# Group the data by date, employee, and aisle and calculate the count of non-null values
Report_By_Aisle1 = pd.pivot_table(data, index='User', values='Qty', columns=['Date', 'Aisle'],aggfunc=sum, fill_value='')
Report_By_Aisle1 = Report_By_Aisle1.applymap(lambda x: abs(x) if pd.notnull(x) and isinstance(x, (int, float)) else x)

# Remove decimal places and format as integers
Report_By_Aisle1 = Report_By_Aisle1.applymap(lambda x: int(x) if pd.notnull(x) and isinstance(x, float) else x)

Report_By_Aisle1 = Report_By_Aisle1.style

# Apply CSS styles to bold and center align the row and column headings
Report_By_Aisle1.set_table_styles([
    {'selector': 'th',
     'props': [
         ('text-align', 'center'),
         ('font-weight', 'bold')
     ]},
    {'selector': 'td',
     'props': [
         ('text-align', 'center')
     ]}
])


# Extract the underlying DataFrames from Styler objects
Report_By_Aisle_df = Report_By_Aisle1.data
Report_By_Employee_df = Report_By_Employee.data

# Concatenate the two pivot tables by date
Report_By_Aisle = pd.concat([Report_By_Aisle_df, Report_By_Employee_df], axis=1)

print("\033[1mPicking Productivity Report By Aisle \033[0m")
print("------------------------------------")
display(Report_By_Aisle)
```

## Export to Excel

```python
# Export reports to Excel. 
#Create a Pandas Excel writer using the 'xlsxwriter' engine
writer = pd.ExcelWriter('C:/Users/Sha/Desktop/Yasmine/Kapscomoto/Picking_Productivity_Reports_Yasmine.xlsx', engine='xlsxwriter')



Report_By_Date.to_excel(writer, sheet_name='Report by Date')
Report_By_Employee.to_excel(writer, sheet_name='Report by Employee')
Report_By_Aisle.to_excel(writer, sheet_name='Report by Aisle')

# Save the Excel file
writer.save()
```
